```python

import requests
from bs4 import BeautifulSoup
url="https://www.worldometers.info/coronavirus/?fbclid=IwAR0RYMPhu-J9i5t74sla8KOrJFIhGMs--0YIuvQzoxLeFrvwwr5FLpn42Qw"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
data=soup.find(id="main_table_countries_today")
data
```

```python
heads=data.find_all('th')
heading=[h.get_text().strip() for h in heads]
heading
```

```python
table_body=data.find_all('tr')
table_body
```

```python
table_body=data.find_all('td')
table_body
```

```python
#first=[d.get_text().strip() for d in table_body[0].find_all('td')]
#first
for d in table_body[0].find_all('td'):
    print(d)
```

```python
first=[d.get_text().strip() for d in table_body[0].find_all('td')]
first
```

```python
comp=[]
for t in table_body:
    l=[d.get_text().strip() for d in t.find_all('td')]
    comp.append(l)
len(comp)
```

```python
for c in comp:
    print(c)
```

```python
d={}
for h in heading:
    d[h]=[]
d
```

```python
for k in comp:
    d['#'].append(k[0])
    d['Country,Other'].append((k[1])
    d['TotalCases'].append(k[2])
    d['NewCases'].append(k[3])
    d['TotalDeaths'].append(k[4])
    d['NewDeaths'].append(k[5])
    d['TotalRecovered'].append(k[6])
    d['NewRecovered'].append(k[7])
    d['ActiveCases'].append(k[8])
    d['Serious,Critical'].append(k[9])
    d['Tot\xa0Cases/1M pop'].append(k[10])
    d[ 'Deaths/1M pop'].append(k[11])
    d['TotalTests'].append(k[12])
    d['Tests/\n1M pop'].append(k[13])
    d['Population'].append(k[14])
    d['Continent'].append(k[15])
    d[ '1 Caseevery X ppl'].append(k[16])
    d['1 Deathevery X ppl'].append(k[17])
    d['1 Testevery X ppl'].append(k[18])
    d['New Cases/1M pop'].append(k[19])
    d['New Deaths/1M pop'].append(k[20])
    d['Active Cases/1M pop'].append(k[21])
    
        
```

```python
import pandas as pd
df=pd.DataFrame(d)
df.head(10)
```

```python
df.to_csv('Reported_Cases_and_Deaths_by_Country_or_Territory.csv',index=False)
```
