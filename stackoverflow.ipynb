```python
import requests
from bs4 import BeautifulSoup
url="https://stackoverflow.com/questions/tagged/python?tab=newest&page=1&pagesize=15"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
all_question=soup.find_all(class_="mln24")
(all_question)
```

```python
first=all_question[0]
first
```

```python
fq=first.find(class_="question-hyperlink").get_text().strip().replace('\n',' ')
fq
```

```python
firstq=[]
for f in all_question:
    firstq.append(f.find(class_="question-hyperlink").get_text().strip().replace('\n',' '))
    
len(firstq)
```

```python
import pandas as pd
d={'firstq':[]}
for i in range(1,7):
    url="https://stackoverflow.com/questions/tagged/python?tab=newest&page={i}&pagesize=15"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html5lib')
    all_question=soup.find_all(class_="mln24")
    for f in all_question:
        d['firstq'].append(f.find(class_="question-hyperlink").get_text().strip().replace('\n',' '))
df=pd.DataFrame(d)
df.head(5)
```

```python
for d in all_question:
    firstq.append(d.find(class_="question-hyperlink").get_text().strip().replace('\n',' '))
    d={"question":firstq}
d

import pandas as pd
df=pd.DataFrame(d)
df
    
```

```python
df.to_csv("all_question.csv",index=False)
```
